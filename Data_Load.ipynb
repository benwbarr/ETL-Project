```python
# Loading Economyu csv data created from Transformation Activity
eco_data_csv_path = 'Resources/Economy_Data.csv'
eco_df = pd.read_csv(eco_data_csv_path)
eco_df.head()
```

```python
eco_df = eco_df.drop(columns=['Unnamed: 0'])
eco_df
```

```python
# Loading Health csv data created from Transformation Activity
health_data_csv_path = 'Resources/Health_Data.csv'
health_df = pd.read_csv(health_data_csv_path)
health_df.head()
```

```python
health_df = health_df.drop(columns=['Unnamed: 0'])
health_df
```

```python
eco_dict = {}
eco_data = {}
for index, row in eco_df.iterrows():
    eco_data = {'GDP Ranking': row['GDP Ranking']
                ,'GDP': row['GDP']
                ,'Ease of Doing Business Rank': row['Ease of Doing Business Rank']
                ,'Starting a Business': row['Starting a Business']
                ,'Dealing with Construction Permits': row['Dealing with Construction Permits']
                ,'Getting Electricity': row['Getting Electricity']
                ,'Registering Property': row['Registering Property']
                ,'Getting Credit': row['Getting Credit']
                ,'Protecting Minority Investors': row['Protecting Minority Investors']
                ,'Paying Taxes': row['Paying Taxes']
                ,'Trading across Borders': row['Trading across Borders']
                ,'Enforcing Contracts': row['Enforcing Contracts']
                ,'Resolving Insolvency': row['Resolving Insolvency']
               }
    eco_dict.setdefault('Country',[]).append(row['Country'])
    eco_dict.setdefault('Economy',[]).append(eco_data)

# Storing Dictionary to DataFrame for easy conversion to Json
eco_final_df = pd.DataFrame(eco_dict)
eco_final_df.head()
```

```python
# Preparing Health Data
health_dict = {}
health_data = {}

for index, row in health_df.iterrows():
    health_data = {'Covid Confirmed': row['Covid Confirmed']
                ,'Covid Deaths': row['Covid Deaths']
                ,'Covid Recovered': row['Covid Recovered']
                ,'Covid first Date': row['Covid first Date']
                ,'Flu/Pneumonia Death Rate Per 100000': row['Flu/Pneumonia Death Rate Per 100000']
               }
    health_dict.setdefault('Country',[]).append(row['Country'])
    health_dict.setdefault('Health',[]).append(health_data)

# Storing Dictionary to DataFrame for easy conversion to Json
health_final_df = pd.DataFrame(health_dict)
health_final_df.head()
```

```python
# Preparing Health Data
health_dict = {}
health_data = {}

for index, row in health_df.iterrows():
    health_data = {'Covid Confirmed': row['Covid Confirmed']
                ,'Covid Deaths': row['Covid Deaths']
                ,'Covid Recovered': row['Covid Recovered']
                ,'Covid first Date': row['Covid first Date']
                ,'Flu/Pneumonia Death Rate Per 100000': row['Flu/Pneumonia Death Rate Per 100000']
               }
    health_dict.setdefault('Country',[]).append(row['Country'])
    health_dict.setdefault('Health',[]).append(health_data)

# Storing Dictionary to DataFrame for easy conversion to Json
health_final_df = pd.DataFrame(health_dict)
health_final_df.head()
```

```python
# Merging all Dataframes to create 1 Dataframe for upload to Mongo DB
final_df = pd.merge(country_df,eco_final_df,on="Country")
final_df = pd.merge(final_df,health_final_df,on="Country")
final_df
```

```python
# Converting Final dataframe to Json for loading to Mongo DB Collection
final_json = json.loads(final_df.to_json(orient='records'))
final_json
```

### Data Load Section

```python
# The default port used by MongoDB is 27017
# https://docs.mongodb.com/manual/reference/default-mongodb-port/
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Defining the 'etlDB' database in Mongo
db = client.etlDB

# Defining collections in the DB
cntry_coll = db.CountryData
```
