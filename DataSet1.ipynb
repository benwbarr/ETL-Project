```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
Rankings_file = "Resources/Rankings.csv"
Rankings_df = pd.read_csv(Rankings_file)
Rankings_df.head()
```

```python
GDP_file = "Resources/GDP.csv"
GDP_df = pd.read_csv(GDP_file, encoding="ISO-8859-1")
GDP_df.head()
```

```python
Growth_file = "Resources/Growth.csv"
Growth_df = pd.read_csv(Growth_file)
Growth_df.head()
```

```python
Health_file = "Resources/Health.csv"
Health_df = pd.read_csv(Health_file)
Health_df.head()
```

```python
Final_file = "Resources/final.csv"
Final_Merge = pd.read_csv(Final_file)
Final_Merge.head()
```

```python
GDP_drop = GDP_df.drop(GDP_df.columns[[0,2,5]], axis = 1)
GDP_drop.head()
```

```python
GDP_final= GDP_drop.rename(columns={"Ranking": "GDP Ranking",
                                    "Economy": "Country",
                                    "US dollars)": "GDP"})
GDP_final.head()
```

```python
GDP_final.set_index("Country", inplace=True)
GDP_final.head()
```

```python
Rankings_final = Rankings_df.rename(columns = {"Economy": "Country"})
Rankings_final.head()
```

```python
Rankings_final.set_index("Country", inplace=True)
Rankings_final.head()
```

```python
Growth_drop = Growth_df.drop(Growth_df.columns[[4,6,7,8,9,11,12,13,14,15,16,17]], axis = 1)
Growth_drop.head()
```

```python
Growth_final = Growth_drop.rename(columns = {"Country": "Country",
                                             "World Percentage": "World Percentage of Population",})
                                             
Growth_final.head()
```

```python
Growth_final.set_index("Country", inplace=True)
Growth_final.head()
```

```python
Health_final = Health_df.rename(columns = {"Rank": "Helath Care Rank",
                                           "Health_Care_Index": "Health Care Approval Rating"})
Health_final.head()
```

```python
Health_final.set_index("Country", inplace=True)
Health_final.head()
```

```python
merge1 = pd.merge(GDP_final,Rankings_final, on = ["Country"])
merge2 = pd.merge(merge1,Growth_final, on = ["Country"])
merge_final = pd.merge(merge2,Health_final, on = ["Country"])

merge_final.head()
```

```python
Final = pd.merge(merge_final,Final_Merge, on = ["Country"])
Final.head()
```

```python
Final.info
```

```python
Final.to_csv("Resources\Mongo_Data.csv")
```
